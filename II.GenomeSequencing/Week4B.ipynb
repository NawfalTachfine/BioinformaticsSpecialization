## Spectral Convolution Problem

```python
from itertools import combinations

def convolution(spectrum):
    diffs = [abs(a-b) for a,b in combinations(sorted(spectrum), 2)]
    return [d for d in diffs if d != 0]
```

```python
def convolution_wrapper(spectrum):
    spectrum_ = [int(s) for s in spectrum.split(' ')]
    convolutions = convolution(spectrum_)
    return ' '.join([str(i) for i in convolutions])
```

```python
sample_input = '0 137 186 323'
sample_output = '137 137 186 186 323 49'
```

```python
assert sorted(convolution_wrapper(sample_input).split(' ')) == sorted(sample_output.split(' '))
```

```python
from collections import Counter
```

```python
def convolution_cyclopeptide_sequencing(
    spectrum,
    convolutions_size, 
    leaderboard_size, 
):
    convolutions_ = [
        mass 
        for mass in convolution(spectrum)
        if 57 <= mass <= 200
    ]
    counter = Counter(convolutions_)
    ordered_counts = counter.most_common()
    cutoff = ordered_counts[convolutions_size - 1][1]
    integer_masses = {
        chr(mass): mass 
        for mass, frequency in counter.items() 
        if frequency >= cutoff
    }
    
    _, leaderboard = leaderboard_cyclopeptide_sequencing(
        spectrum, leaderboard_size, integer_masses)
    return leaderboard
```

```python
sample_spectrum_ = '57 57 71 99 129 137 170 186 194 208 228 265 285 299 307 323 356 364 394 422 493'
sample_spectum = parse_spectrum(sample_spectrum_)
sample_m = 20
sample_n = 60
sample_output = '99-71-137-57-72-57'
```

```python
sample_peptides = convolution_cyclopeptide_sequencing(sample_spectum, sample_m, sample_n)
assert sample_output in sample_peptides
```
